```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings(action='ignore') 
```

```python
data_path = ''
```

# 예측모델 결과 활용하여 피처 생성

```python
train_4 = pd.read_csv(data_path + 't1.csv')
train_4.head()
```

```python
train_5 = pd.read_csv(data_path + 't2.csv')
train_5.head()
```

```python
test_6 = pd.read_csv(data_path + 't3.csv')
test_6.head()
```

## 예측 모델링 단계에서 중요 피쳐로 뽑힌 피처만 사용하여 새로운 피처 만들기

```python
important_col = ['event_count', 'loanapply_insert_weekday', 'loanapply_insert_hour',
       'existing_loan_cnt', 'company_enter_month', 'credit_score_f',
       'product_applied', 'credit_score', 'desired_amount_rate',
       'loan_rate', 'most_week', 'continuous_use', 'bank_applied',
       'over_credit_interest', 'over_interest', 'line_app_applied',
       'event_nunique', 'purpose_생활비', 'desired_amount_YN',
       'desired_amount', 'loan_limit', 'have_c', 'visit_count', 'have_e',
       'time_days', 'yearly_income', 'st_interest_rate', 'st_interest',
       'have_b', 'birth_year', 'purpose_대환대출', 'loanapply_insert_ohu']
```

```python
# 추출할 칼럼 리스트 생성
# is_applied, application_id, product_id, user_id
important_col.append('is_applied')
important_col.append('application_id')
important_col.append('product_id')
important_col.append('user_id')
```

```python
train_4 = train_4[important_col]
train_5 = train_5[important_col]
test_6 = test_6[important_col]
```

```python
# 예측 모델 결과에 test_6 피처 붙이기
test_6_result = pd.read_csv(data_path + '데이터분석분야_퓨처스부문_EZ수고_평가데이터.csv')
test_6_result.head()
```

```python
test_6_ = pd.merge(test_6_result, test_6, how = 'left', on = ['application_id', 'product_id'])
```

```python
del test_6_['is_applied_y']
test_6_.rename(columns = {'is_applied_x' : 'is_applied'}, inplace = True)
```

```python
del test_6
del test_6_result
```

```python
# 4~6월 예측 모델링 피처 및 결과 합치기
features = pd.concat([train_4, train_5, test_6_])
features.head()
```

```python
features.user_id.nunique()
```

## 예측모델 features, user_id로 groupby 진행
기존에 이미 전처리해둔 모델이기 때문에 모두 mean 값 사용하여 groupby 진행

```python
user_col = ['is_applied', 'application_id', 'product_id', 'user_id']
for x in user_col :
    important_col.remove(x)
```

```python
features[important_col]
```

```python
user = pd.DataFrame({'user_id' : features.user_id.unique()})
```

```python
for x in important_col :
    f = features[['user_id', x]].groupby('user_id')[x].mean().reset_index()
    user = pd.merge(user, f, how = 'left', on = 'user_id')
    del f
```

```python
user
```

## is_applied를 활용한 피처 생성

- is_applied == 1 비율

```python
f = features.groupby('application_id')['is_applied'].mean().reset_index()
```

```python
f = pd.merge(f, features[['application_id', 'user_id']], how = 'left', on = 'application_id')
f = f.groupby('user_id')['is_applied'].mean().reset_index()
f
```

```python
user = pd.merge(user, f, how = 'left', on = 'user_id')
user
```

```python
user.rename(columns={'is_applied' : 'mean_applied'}, inplace=True)
```

```python
del f
```

- is_applied == 1인 product_id 개수

```python
product_score = features[features['is_applied'] == 1]['product_id'].value_counts().reset_index()
product_score.rename(columns = {'index' : 'product_id', 'product_id' : 'applied_cnt'}, inplace = True)
product_score['rank'] = product_score['applied_cnt'].rank(method='dense', ascending = True)
product_score
```

```python
product = {product_score.loc[i, 'product_id'] : product_score.loc[i, 'rank'] for i in range(len(product_score))}
```

```python
f = features[features['is_applied'] == 1][['user_id', 'product_id']]
f
```

```python
f['applied_product_score'] = f['product_id'].apply(lambda x : product[x])
```

```python
f = f.groupby('user_id')['applied_product_score'].mean().reset_index()
```

```python
f
```

```python
user = pd.merge(user, f, how = 'left', on = 'user_id', )
user
del f
```

```python
# applied_product_score가 없으면 applied 한 경우가 없는 경우로 0으로 결측치 채워줌
user['applied_product_score'].fillna(0, inplace = True)
```

# 추가로 넣어볼 피쳐

```python
# 대출과 관련된 정보(경제 정보)를 얼마나 알고있는지 여부 또한 중요할 것으로 생각
# 이를 자신의 경제 상황을 얼마나 알고있는지로 확인 (제공 데이터 중 user_spec의 응답여부로 판단)

user_spec = pd.read_csv(data_path + 'user_spec.csv')
user_spec.head()
```

```python
user_spec = user_spec.query("user_id in @user.user_id.unique()")
```

## 결측 응답 개수 구하기 `must_col_no_ans`

### 필수 정보 칼럼
- application_id 별 필수정보 결측치 있는 개수
- 필수정보칼럼 : birth_year, gender, credit_score, yearly_income, income_type, houseown_type, desired_amount, purpose, personal_rehabilitation_yn

```python
must_col = ['birth_year', 'gender', 'credit_score', 'yearly_income', 'income_type', 'houseown_type', 'desired_amount', 'purpose', 'personal_rehabilitation_yn']
user_spec['must_col_no_ans'] = user_spec.loc[:, must_col].isna().sum(axis = 'columns')
```

### 연관 정보 칼럼

**[personal_rehabilitation_yn, personal_rehabilitation_complete_yn]**

```python
# personal_rehabilitation_yn
# personal_rehabilitation_complete_yn

# personal_rehabilitation_yn = 1 (개인회생자)면
# personal_rehabilitation_complete_yn 있어야함

user_spec[user_spec['personal_rehabilitation_yn'] == 1]['personal_rehabilitation_complete_yn'].isna().sum()
```

```python
# personal_rehabilitation_complete_yn가 있는데
# personal_rehabilitation_yn가 없는 애들

user_spec[user_spec['personal_rehabilitation_complete_yn'].isna() == False][user_spec['personal_rehabilitation_yn'].isna()]
```

**[existing_loan_cnt, existing_loan_amt]**

```python
# existing_loan_cnt 없는데 existing_loan_amt 있는 사람
user_spec[user_spec['existing_loan_cnt'].isna()][user_spec['existing_loan_amt'].isna() == False]
```

```python
# existing_loan_amt 없는데 existing_loan_cnt 있는 사람
user_spec[user_spec['existing_loan_amt'].isna()][user_spec['existing_loan_cnt'].isna() == False]
```

```python
wrong_ans_users_index = user_spec[user_spec['existing_loan_amt'].isna()][user_spec['existing_loan_cnt'].isna() == False].index
user_spec.loc[wrong_ans_users_index, 'must_col_no_ans'] += 1
del wrong_ans_users_index
```

**[employment_type, company_enter_month]**

```python
user_spec['employment_type'].unique()
```

```python
# 정규직이거나 계약직인데 company_enter_month가 없는 경우
wrong_ans_users_index = user_spec.query("employment_type in ['정규직', '계약직']")[user_spec['company_enter_month'].isna()].index
user_spec.loc[wrong_ans_users_index, 'must_col_no_ans'] += 1
del wrong_ans_users_index
```

```python
# employment_type이 nan 값인데 company_enter_month가 있는 경우
wrong_ans_users_index = user_spec[user_spec['employment_type'].isna()][user_spec['company_enter_month'].isna() == False].index
user_spec.loc[wrong_ans_users_index, 'must_col_no_ans'] += 1
del wrong_ans_users_index
```

```python
f = user_spec.groupby('user_id')['must_col_no_ans'].mean().reset_index()
f
```

```python
user = pd.merge(user, f, how = 'left', on = 'user_id')
del f
user
```
